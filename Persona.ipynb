```python
import re
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from PyPDF2 import PdfWriter, PdfReader
from io import BytesIO

# Function to parse the markdown content
def parse_markdown(markdown_text):
    data = {}
    
    data['ชื่อ-นามสกุล'] = re.search(r'- ชื่อ-นามสกุล: `<(.+?)>`', markdown_text).group(1)
    data['ชื่อเล่น'] = re.search(r'- ชื่อเล่น: `<(.+?)>`', markdown_text).group(1)
    data['รหัสนักศึกษา'] = re.search(r'- รหัสนักศึกษา: `<(.+?)>`', markdown_text).group(1)
    
    data['ความถนัด'] = []
    if re.search(r'\[x\] ทางกล', markdown_text) :
        data['ความถนัด'].append('ทางกล')
    if re.search(r'\[x\] ไฟฟ้า', markdown_text):
        data['ความถนัด'].append('ไฟฟ้า')
    if re.search(r'\[x\] เขียนโปรแกรม', markdown_text):
        data['ความถนัด'].append('เขียนโปรแกรม')
    if re.search(r'\[x\] คุมโปรเจค', markdown_text):
        data['ความถนัด'].append('คุมโปรเจค')
    
    data['สิ่งที่ชอบ'] = []
    if re.search(r'\[x\] ทางกล', markdown_text):
        data['สิ่งที่ชอบ'].append('ทางกล')
    if re.search(r'\[x\] ไฟฟ้า', markdown_text):
        data['สิ่งที่ชอบ'].append('ไฟฟ้า')
    if re.search(r'\[x\] เขียนโปรแกรม', markdown_text):
        data['สิ่งที่ชอบ'].append('เขียนโปรแกรม')
    if re.search(r'\[x\] คุมโปรเจค', markdown_text):
        data['สิ่งที่ชอบ'].append('คุมโปรเจค')
    
    data['เป้าหมาย'] = re.search(r'- เป้าหมายที่อยากเรียนรู้: `<(.+?)>`', markdown_text).group(1)
    data['อุปนิสัย'] = re.search(r'- อุปนิสัย: `<(.+?)>`', markdown_text).group(1)
    data['ประสบการณ์'] = re.search(r'- ประสบการณ์: `<(.+?)>`', markdown_text).group(1)
    data['โปรเจค'] = re.search(r'- โปรเจค/นวัตกรรมเพื่อสังคมที่อยากจะทำ \(ถ้ามี\): `<(.+?)>`', markdown_text).group(1)
    
    return data

# Function to create the filled PDF
def create_filled_pdf(data, template_path, output_path):
    packet = BytesIO()
    can = canvas.Canvas(packet, pagesize=letter)
    
    # Fill in the data
    can.drawString(100, 750, f"ชื่อ-นามสกุล: {data['ชื่อ-นามสกุล']}")
    can.drawString(100, 730, f"ชื่อเล่น: {data['ชื่อเล่น']}")
    can.drawString(100, 710, f"รหัสนักศึกษา: {data['รหัสนักศึกษา']}")
    
    # Add the "check" marks for ความถนัด
    y_pos = 690
    for skill in data['ความถนัด']:
        can.drawString(100, y_pos, f"- [x] {skill}")
        y_pos -= 20
    
    # Add the "check" marks for สิ่งที่ชอบ
    y_pos = y_pos - 20
    for interest in data['สิ่งที่ชอบ']:
        can.drawString(100, y_pos, f"- [x] {interest}")
        y_pos -= 20
    
    can.drawString(100, y_pos - 20, f"เป้าหมายที่อยากเรียนรู้: {data['เป้าหมาย']}")
    can.drawString(100, y_pos - 40, f"อุปนิสัย: {data['อุปนิสัย']}")
    can.drawString(100, y_pos - 60, f"ประสบการณ์: {data['ประสบการณ์']}")
    can.drawString(100, y_pos - 80, f"โปรเจค/นวัตกรรมเพื่อสังคมที่อยากจะทำ: {data['โปรเจค']}")
    
    can.save()
    
    # Move to the beginning of the StringIO buffer
    packet.seek(0)
    new_pdf = PdfReader(packet)
    
    # Read the existing PDF
    existing_pdf = PdfReader(open(template_path, "rb"))
    output = PdfWriter()
    
    # Add the "watermark" (which is the new pdf) on the existing page
    page = existing_pdf.pages[0]
    page.merge_page(new_pdf.pages[0])
    output.add_page(page)
    
    # Finally, write "output" to a real file
    with open(output_path, "wb") as outputStream:
        output.write(outputStream)



# Reading the markdown file
markdown_file_path = './111ncxyz.md'

with open(markdown_file_path, 'r', encoding='utf-8') as file:
    markdown_content = file.read()

# Parse the markdown content
data = parse_markdown(markdown_content)
print(data)
# Paths to the template and output file
template_path = './FRAB11-Self-Introduction%.pdf'
output_path = './filled_info/output_filled.pdf'

# Create the filled PDF
create_filled_pdf(data, template_path, output_path)
```

```python
import os
import re
from PIL import Image, ImageDraw, ImageFont

# Function to parse the markdown content
def wrap_text(text, width=50):
    """Wrap text to the specified width."""
    return '\n'.join([text[i:i+width] for i in range(0, len(text), width)])

import re
import pythainlp

def wrap_text(text, max_length):
    words = pythainlp.tokenize.word_tokenize(text)  # ตัดคำ
    lines = []
    current_line = ""
    
    for word in words:
        if len(current_line) + len(word) + 1 <= max_length:
            if current_line:
                current_line += ""
            current_line += word
        else:
            lines.append(current_line)
            current_line = word
    
    if current_line:
        lines.append(current_line)
    
    return "\n".join(lines)

def parse_markdown(markdown_text):
    data = {}

    # Extract 'ความถนัด' section content
    khamtan_match = re.search(r'- ความถนัด \(เลือกได้มากกว่า 1 ข้อ\)(.*?)\n- ', markdown_text, re.DOTALL)
    if khamtan_match:
        khamtan_content = khamtan_match.group(1)
        data['ความถนัด'] = []
        if re.search(r'\[x\] ทางกล', khamtan_content) or re.search(r'\[X\] ทางกล', khamtan_content):
            data['ความถนัด'].append('ทางกล')
        if re.search(r'\[x\] ไฟฟ้า', khamtan_content) or re.search(r'\[X\] ไฟฟ้า', khamtan_content):
            data['ความถนัด'].append('ไฟฟ้า')
        if re.search(r'\[x\] เขียนโปรแกรม', khamtan_content) or re.search(r'\[X\] เขียนโปรแกรม', khamtan_content):
            data['ความถนัด'].append('เขียนโปรแกรม')
        if re.search(r'\[x\] คุมโปรเจค', khamtan_content) or re.search(r'\[X\] คุมโปรเจค', khamtan_content):
            data['ความถนัด'].append('คุมโปรเจค')

        if re.search(r'\[x\] อื่นๆ', khamtan_content) or re.search(r'\[X\] อื่นๆ', khamtan_content):
            data['ความถนัด'].append('อื่นๆ')
            
            # Check for specific content after 'อื่นๆ'
            # Handle the 'อื่นๆ' case
            other_match = re.search(r'\[x\] อื่นๆ\s*(.*?)\n', khamtan_content, re.DOTALL | re.IGNORECASE)
            if other_match:
                other_content = other_match.group(1).strip()
                if other_content:
                    data['ความถนัด'].append(f'อื่นๆ: {other_content}')

    # Extract 'สิ่งที่ชอบ/อยากพัฒนา' section content
    singchob_match = re.search(r'- สิ่งที่ชอบ/อยากพัฒนา \(เลือกได้มากกว่า 1 ข้อ\)(.*?)\n- ', markdown_text, re.DOTALL)
    if singchob_match:
        singchob_content = singchob_match.group(1)
        data['สิ่งที่ชอบ'] = []
        if re.search(r'\[x\] ทางกล', singchob_content) or re.search(r'\[X\] ทางกล', singchob_content):
            data['สิ่งที่ชอบ'].append('ทางกล')
        if re.search(r'\[x\] ไฟฟ้า', singchob_content) or re.search(r'\[X\] ไฟฟ้า', singchob_content):
            data['สิ่งที่ชอบ'].append('ไฟฟ้า')
        if re.search(r'\[x\] เขียนโปรแกรม', singchob_content) or re.search(r'\[X\] เขียนโปรแกรม', singchob_content):
            data['สิ่งที่ชอบ'].append('เขียนโปรแกรม')
        if re.search(r'\[x\] คุมโปรเจค', singchob_content) or re.search(r'\[X\] คุมโปรเจค', singchob_content):
            data['สิ่งที่ชอบ'].append('คุมโปรเจค')

        if re.search(r'\[x\] อื่นๆ', singchob_content) or re.search(r'\[X\] อื่นๆ', singchob_content):
            data['สิ่งที่ชอบ'].append('อื่นๆ')
            
            # Check for specific content after 'อื่นๆ'
         # Handle the 'อื่นๆ' case
            other_match = re.search(r'\[x\] อื่นๆ\s*(.*?)\n', singchob_content, re.DOTALL | re.IGNORECASE)
            if other_match:
                other_content = other_match.group(1).strip()
                if other_content:
                    data['สิ่งที่ชอบ'].append(f'อื่นๆ: {other_content}')

    # Extracting other sections
    
    

    goal_match = re.search(r'- เป้าหมายที่อยากเรียนรู้: `<(.+?)>`', markdown_text)
    goal_text = goal_match.group(1) if goal_match else ''
    if len(goal_text) > 250:
        data['เป้าหมาย'] = wrap_text(goal_text, 75)
    else:
        data['เป้าหมาย'] = wrap_text(goal_text, 59)



    # Extracting อุปนิสัย
    habit_match = re.search(r'- อุปนิสัย: `<(.+?)>`', markdown_text)
    habit_text = habit_match.group(1) if habit_match else ''
    if len(habit_text) > 130:
         data['อุปนิสัย'] = wrap_text(habit_text, 75)
    else:
        data['อุปนิสัย'] = wrap_text(habit_text, 65)

    # Extracting ประสบการณ์
    experience_match = re.search(r'- ประสบการณ์: `<(.+?)>`', markdown_text)
    experience_text = experience_match.group(1) if experience_match else ''
    if len(experience_text) > 220:
        data['ประสบการณ์'] = wrap_text(experience_text, 75)
    else:
        data['ประสบการณ์'] = wrap_text(experience_text, 60)
    # Extracting โปรเจค
    project_match = re.search(r'- โปรเจค/นวัตกรรมเพื่อสังคมที่อยากจะทำ \(ถ้ามี\): `<(.+?)>`', markdown_text)
    project_text = project_match.group(1) if project_match else ''
    if len(project_text) > 220:
        data['โปรเจค'] = wrap_text(project_text, 75)
    else:
        data['โปรเจค'] = wrap_text(project_text, 55)
    
    return data

# Function to add text and "x" marks to an image
def add_text_to_image(image_path, output_path, data):
    # Load the image
    image = Image.open(image_path)
    draw = ImageDraw.Draw(image)
    
    # Load a TTF font
    front_size = 35
    front_size_small = 25
    font = ImageFont.truetype("./Prompt Regular.ttf", front_size)  # Adjust the font size as needed
    x_font = ImageFont.truetype("./Prompt Regular.ttf", front_size)  # Slightly larger font for the "x"
    other_front = ImageFont.truetype("./Prompt Regular.ttf", front_size_small)

    if len(data['เป้าหมาย']) > 250:
        font = ImageFont.truetype("./Prompt Regular.ttf", front_size_small)
    if len(data['อุปนิสัย']) > 150:
        font = ImageFont.truetype("./Prompt Regular.ttf", front_size_small)
    if len(data['ประสบการณ์']) > 220:
        font = ImageFont.truetype("./Prompt Regular.ttf", front_size_small)
    
    # Set text position and color
    text_color = (0, 0, 139)  # Dark blue color
    x_color = (0, 0, 139)  # Dark blue color for the "x"
    
    # Coordinates for the boxes where "x" should be placed
    box_coordinates = {
        'ทางกล': (930, 115),  # Adjust these coordinates as needed
        'ไฟฟ้า': (1090, 115),  # Adjust these coordinates as needed
        'เขียนโปรแกรม': (1225, 115),  # Adjust these coordinates as needed
        'คุมโปรเจค': (1475, 115)  # Adjust these coordinates as needed
    }

    box_coordinate_2 = {
        'ทางกล': (1090, 200),  # Adjust these coordinates as needed
        'ไฟฟ้า': (1245, 200),  # Adjust these coordinates as needed
        'เขียนโปรแกรม': (1380, 200),  # Adjust these coordinates as needed
        'คุมโปรเจค': (1635, 200)  # Adjust these coordinates as needed
    }

    # Draw the "x" in the boxes for the selected skills
    for skill in data['ความถนัด']:
        if skill in box_coordinates:
            draw.text(box_coordinates[skill], "x", font=x_font, fill=x_color)
        elif skill.startswith("อื่นๆ"):
            # Handle custom 'อื่นๆ' content
            other_text_position = (930, 160)
            clear_skill = skill.replace(".", "")
            split_other = clear_skill.split("อื่นๆ:")  
            if len(split_other) > 1:
                other_content = split_other[1].strip()
                draw.text(other_text_position,other_content, font=other_front, fill=text_color)


    for skill in data['สิ่งที่ชอบ']:
        if skill in box_coordinate_2:
            draw.text(box_coordinate_2[skill], "x", font=x_font, fill=x_color)
        elif skill.startswith("อื่นๆ"):
            # Handle custom 'อื่นๆ' content
            other_text_position = (930, 250)  # Adjust the coordinates as needed
            # delected . ใน skill
            clear_skill = skill.replace(".", "")
            split_other = clear_skill.split("อื่นๆ:")  
            if len(split_other) > 1:
                other_content = split_other[1].strip()
                draw.text(other_text_position,other_content, font=other_front, fill=text_color)

    # Define the starting position for additional text
    x_position = 800  # Starting x position for the text
    y_position = 320  # Starting y position for the text

    y_position += 20  # Add some space

    # Draw the remaining data
    draw.text((x_position, y_position), f"{data['เป้าหมาย']}", font=font, fill=text_color)
    y_position += 200
    draw.text((x_position, y_position), f"{data['อุปนิสัย']}", font=font, fill=text_color)
    y_position += 150
    draw.text((x_position, y_position), f"{data['ประสบการณ์']}", font=font, fill=text_color)
    y_position += 170
    draw.text((x_position, y_position), f"{data['โปรเจค']}", font=font, fill=text_color)

    # Save the image with the added text
    image.save(output_path)


# Loop through files in the directories
image_dir = './Template_image'
markdown_dir = './Persona_67_submission'
output_dir = './filled_info'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for filename in os.listdir(image_dir):
    if filename.endswith(".jpg"):
        image_path = os.path.join(image_dir, filename)
        markdown_file_path = os.path.join(markdown_dir, filename.replace(".jpg", ".md"))
        output_path = os.path.join(output_dir, f"{filename.replace('.jpg', '_filled.jpg')}")

        if os.path.exists(markdown_file_path):
            with open(markdown_file_path, 'r', encoding='utf-8') as file:
                markdown_content = file.read()
            
            # Parse the markdown content
            
            data = parse_markdown(markdown_content)
            
            # Add text and "x" marks to the image
            add_text_to_image(image_path, output_path, data)
```
